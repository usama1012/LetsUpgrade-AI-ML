```python
import pandas as pd
```

```python
import numpy as np
```

```python
from sklearn import tree
```

```python
from sklearn import preprocessing
```

```python
titanic_train=pd.read_csv("train.csv")
```

```python
titanic_train["Age"].mean()
```

```python
meanage=np.where(titanic_train["Age"].isnull(),32,titanic_train["Age"])
```

```python
titanic_train["Age"]=meanage
```

```python
label_encoder=preprocessing.LabelEncoder()
```

```python
encode_sex=label_encoder.fit_transform(titanic_train["Sex"])
```

```python
tree_model=tree.DecisionTreeClassifier()
```

```python
predictors=pd.DataFrame([encode_sex,titanic_train["Age"],titanic_train["Fare"]]).T
```

```python
tree_model.fit(X=predictors,y=titanic_train["Survived"])
```

```python
with open("Titanic.dot",'w') as f:
    f=tree.export_graphviz(tree_model,feature_names=["Sex","Age","Fare"],out_file=f)
```

```python
tree_model.score(X=predictors,y=titanic_train["Survived"])
```

### The model is 97.6% Accurate and Hence can be used to decide which all pasengers survived and which did not
